```python
##LABEL ENCODE
##convert the data into machine readable form=numeric
##species atribute is categorical
```

```python
 from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['species'] =le.fit_transform(df['species'])
df.head()
```

```python
##MODEL TRAINING

#train =70%
#test =30%
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns= ['species'])   #input attribute
y = df['species']  #output attribute
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.3)
```

```python
 #logistic regression   -classification model

from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
```

```python
##MODEL TRAINING

lr_model.fit(x_train, y_train) 
```

```python
print("Accuracy  [LinearRegresion]= ", lr_model.score(x_test, y_test)*100)
```

```python
##KNN= k-nearesr neighbours

from sklearn.neighbors import KNeighborsClassifier
k_model = KNeighborsClassifier()
```

```python
k_model.fit(x_train, y_train)
```

```python
print("Accuracy  [KNN]= ", k_model.score(x_test, y_test)*100)
```

```python
##DEcision tree

from sklearn.tree import DecisionTreeClassifier
d_model = DecisionTreeClassifier()
```

```python
d_model.fit(x_train, y_train)
```

```python
print("Accuracy  [DECISION TREE]= ", d_model.score(x_test, y_test)*100)
```
